```python
# パッケージのインポート
import tweepy
import time
```

```python
# TwitterAPIKey(https://developer.twitter.com/en/portal/projects-and-apps)
CONSUMER_KEY = 'qQep5FCPaJ3l7De5Ia5NvXFX1'
CONSUMER_SECRET = 'kQoQb0Bl6Jap3iE6wudFyyK678h2nrE9RwNHFCLevP53FQ650f'
ACCESS_TOKEN = '1470940921051369475-oTjQNNCRci8MGF1MEizK9JFVRYc65r'
ACCESS_SECRET = '75BUve4v18qb5TFObvHGiVcSiZKB55dUMoUDQwP02Xp6P'

# OAuth認証
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
api = tweepy.API(auth)
```

```python
# followers_ids関数で指定したアカウント（screen_nameにIDを指定）のフォロワーのIDリストを全て取得
results = api.followers_ids(screen_name='jimin_koho')

tweets = ''
start_point = 0
end_point = 800
for i in range(5):
    for result in results[start_point:end_point]:
        try:
            response_list = api.user_timeline(user_id=result)
            for response in response_list:
                tweets += response._json['text']

        except Exception as e:
            print(e)

    with open('jimin_followers_tweets_'+str(start_point)+'-'+str(end_point)+'.txt', mode='a') as f:
        f.write(tweets)

    time.sleep(900)

    start_point = start_point + 800
    end_point = end_point + 800
```

```python
# 形態素解析で出現した候補者名を集計
from janome.tokenizer import Tokenizer
import collections

start_point = 0
end_point = 800
full_text = ''

for i in range(5):
    with open('jimin_followers_tweets_'+str(start_point)+'-'+str(end_point)+'.txt', 'r', encoding='UTF-8') as f:
        full_text = full_text + f.read()

    start_point = start_point + 800
    end_point = end_point + 800

t = Tokenizer()
c = collections.Counter(token.base_form for token in t.tokenize(full_text)
                        if token.part_of_speech.startswith('名詞,固有名詞'))
print(c.most_common()[:30])
```
