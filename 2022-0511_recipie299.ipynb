```python
# 素数の個数を求める
def is_prime(n):
    n_p=0
    for k in range(2,n+1):
        if isPrime(k) :
            n_p=n_p+1
    print(n, " : ", n_p)
            
def isPrime(n):
    num_yakusuu=0
    for i in range(1,n+1):
        if n % i == 0 :
            num_yakusuu = num_yakusuu + 1
    if num_yakusuu==2:
        return True
    else:
        return False
    
is_prime(1000)
```

```python
# n番目に小さい素数を求める
def is_prime2(i):
    if i <= 1:
        return False
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            return False
    return True

def prime_seq(cnt):
    i = 1
    while 0 < cnt:
        i += 1
        if is_prime2(i):
            cnt -= 1
    return i
print(f'{n}番目の素数：',prime_seq(100))
```
