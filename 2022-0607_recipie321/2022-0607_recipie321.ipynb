## 標準ライブラリのmathモジュールを使ってみる

```python
# パッケージのインポート
import math as mt

# 通常の式
n = 3   #引数
m = 1
# m=1に i+1=1,2,3 を順番に乗算していく
for i in range(n):
  m = m*(i+1)
print(m)

# 3! = 6 を計算し表示
print(mt.factorial(3))
```
